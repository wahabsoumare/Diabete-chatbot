```python
urls = [
    'https://www.diabete.qc.ca/le-diabete-en-questions/',
    'https://www.diabete.qc.ca/faq/'
]
my_dict = {}
for url in urls :
  response = requests.get(url)
  if response.status_code == 200 :
      soup = BeautifulSoup(response.content, 'html.parser')
      sections = soup.find_all('section', class_ = 'accordion')
      for section in sections :
          answers = []
          question = section.find('span', class_ = 'accordion__header__title').text
          results = section.find('div', class_ = 'accordion__sub-rows').find_all('p')
          for result in results :
              answers.append(result.text)
          my_dict[question] = ' '.join(answers)

df = pd.DataFrame(my_dict.items(), columns = ['questions', 'answers'])
df
```

```python
# Fichier CSV de sortie
output_file = "diabete_complet.csv"

# Colonnes du fichier CSV
fields = ["Page", "Titre", "Sous-section", "Section", "Contenu", "Liens", "Images"]

# Initialisation du fichier CSV
file = open(output_file, mode="w", newline="", encoding="utf-8")
writer = csv.DictWriter(file, fieldnames=fields)
writer.writeheader()
```

```python
def scrape_page(page_name, page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.find('h1').get_text(strip=True) if soup.find('h1') else "Titre non trouvé"
    return soup, title

def scrape_sections(soup):
    sections = soup.find_all(['h2', 'h3'])
    data = []

    for section in sections:
        section_title = section.get_text(strip=True)
        content = ""
        sibling = section.find_next_sibling()
        while sibling and sibling.name not in ['h2', 'h3']:
            content += sibling.get_text(strip=True) + " "
            sibling = sibling.find_next_sibling()

        links = [a['href'] for a in section.find_all_next('a', href=True)]
        images = [img['src'] for img in soup.find_all('img', src=True)]

        data.append({
            "Section": section_title,
            "Contenu": content.strip(),
            "Liens": "; ".join(links),
            "Images": "; ".join(images)
        })
    return data
```

```python
for page_name, page_data in pages.items():
    print(f"Scraping de la page principale : {page_name}")

    # Scraper la page principale
    soup, title = scrape_page(page_name, page_data["url"])

    # Déterminer les pages à scraper
    urls_to_scrape = page_data["subpages"] if page_data["subpages"] else [page_data["url"]]

    for subpage_url in urls_to_scrape:
        print(f"Scraping de la sous-page : {subpage_url}")
        sub_soup, sub_title = scrape_page(page_name, subpage_url)

        # Scraper les sections
        sections_data = scrape_sections(sub_soup)

        for section_data in sections_data:
            # Écriture dans le fichier CSV
            writer.writerow({
                "Page": page_name,
                "Titre": title,
                "Sous-section": sub_title,
                **section_data  # Ajoute les données de section
            })

file.close()
print(f"Scraping terminé. Les données sont enregistrées dans {output_file}.")https://www.diabete.qc.ca/le-diabete/la-vie-avec-le-diabete/consommation-dalcool-de-tabac-ou-de-drogues/le-cannabis-et-le-diabete/
```

## Pape Malick's one

```python
url = "https://www.diabete.qc.ca/le-diabete/la-gestion-du-diabete/alimentation/"


response = requests.get(url)
response = response.content
soup = BeautifulSoup(response, 'html.parser')
div = soup.find('div', class_='entry-content')


post_items = div.find_all('div', class_='post-item')

all_subpage_data = []

for post_item in post_items:
    link_element = post_item.find('a')
    if link_element:
        subpage_url = link_element['href']

        try:
            subpage_response = requests.get(subpage_url)
            subpage_response.raise_for_status()
            subpage_soup = BeautifulSoup(subpage_response.content, 'html.parser')

            # Data extraction from the subpage
            subpage_title = subpage_soup.find('h1')
            if subpage_title:
                title_text = subpage_title.text.strip()
            else:
                title_text = "No title found"


            all_subpage_data.append({
                "url": subpage_url,
                "title": title_text,
                "body" : subpage_soup.find('div', class_='entry-content').text
            })

        except requests.exceptions.RequestException as e:
            print(f"Error fetching {subpage_url}: {e}")
        except Exception as e:
            print(f"An error occurred processing {subpage_url}: {e}")

df3 = pd.DataFrame(all_subpage_data)
```
